```python
import os
import pycountry
import pycountry_convert as pc
import json

# === 1Ô∏è‚É£ Directorios ===
BASE_DIR = os.path.dirname(os.path.abspath(__file__))  # carpeta donde est√° el script / notebook
RESULTS_DIR = os.path.join(BASE_DIR, '..', 'results')
os.makedirs(RESULTS_DIR, exist_ok=True)

# === 2Ô∏è‚É£ Generar el diccionario ===
iso_to_continent = {}

for country in pycountry.countries:
    try:
        continent_code = pc.country_alpha2_to_continent_code(country.alpha_2)
        continent_name = pc.convert_continent_code_to_continent_name(continent_code)
        # Traducimos al espa√±ol
        traduccion = {
            "Europe": "Europa",
            "Asia": "Asia",
            "Africa": "√Åfrica",
            "Oceania": "Ocean√≠a",
            "North America": "Am√©rica del Norte",
            "South America": "Am√©rica del Sur",
            "Antarctica": "Ant√°rtida"
        }
        iso_to_continent[country.alpha_2] = traduccion.get(continent_name, continent_name)
    except Exception:
        iso_to_continent[country.alpha_2] = "Desconocido"

# === 3Ô∏è‚É£ Guardar en results/ ===
JSON_PATH = os.path.join(RESULTS_DIR, 'iso_to_continent.json')

with open(JSON_PATH, 'w', encoding='utf-8') as f:
    json.dump(iso_to_continent, f, ensure_ascii=False, indent=2)

print(f"‚úÖ Diccionario generado con {len(iso_to_continent)} pa√≠ses.")
print(f"üíæ Guardado en: {JSON_PATH}")

```
